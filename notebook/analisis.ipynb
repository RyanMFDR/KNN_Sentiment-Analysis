### Pembobotan

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
# 1. TF
vectorizer_tf = CountVectorizer()
X_tf = vectorizer_tf.fit_transform(data_cleaned['text'])
tf_df = pd.DataFrame(X_tf.toarray(), columns=vectorizer_tf.get_feature_names_out())

# 2. TF-IDF
vectorizer_tfidf = TfidfVectorizer()
X_tfidf = vectorizer_tfidf.fit_transform(data_cleaned['text'])
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=vectorizer_tfidf.get_feature_names_out())
```

```python
print("TF head:", tf_df.shape)
print("TF-IDF head:", tfidf_df.shape)
```

```python
import matplotlib.pyplot as plt

term_freq = tf_df.sum(axis=0)

top_terms = term_freq.sort_values(ascending=False).head(10)

#Barplot
plt.figure(figsize=(10,6))
top_terms.plot(kind='bar')
plt.title('Top 10 Term Frequency')
plt.ylabel('Frequency')
plt.xlabel('Terms')
plt.show()
```

```python
term_tfidf = tfidf_df.mean(axis=0)

top_tfidf = term_tfidf.sort_values(ascending=False).head(10)

#TF IDF Tertinggi
plt.figure(figsize=(10,6))
top_tfidf.plot(kind='bar', color='orange')
plt.title('Top 10 Terms by Average TF-IDF')
plt.ylabel('Average TF-IDF')
plt.xlabel('Terms')
plt.show()
```

### Selection Features

```python
from sklearn.feature_selection import SelectKBest, chi2

selector = SelectKBest(score_func=chi2, k=5)
X_new = selector.fit_transform(tfidf_df, data_cleaned['label'])

# Get selected feature names
selected_features = tfidf_df.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

```python
#Nilai Chi2
chi2_scores = selector.scores_
selected_mask = selector.get_support()  
selected_features = tfidf_df.columns[selected_mask]
selected_scores = chi2_scores[selected_mask]

result = pd.DataFrame({
    'Feature': selected_features,
    'Chi2 Score': selected_scores
}).sort_values(by='Chi2 Score', ascending=False)

print(result)
```

### Model Awal

```python
#MODEL AWAL
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

x = X_new
y = data_cleaned['label']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(classification_report(y_test, y_pred, target_names=data_cleaned['label'].unique()))
```

```python
#Visualisasi Confusion Matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred, labels=['negative', 'positive'])
import pandas as pd
cm_df = pd.DataFrame(cm, index=['Actual Negative', 'Actual Positive'],
                        columns=['Predicted Negative', 'Predicted Positive'])

plt.figure(figsize=(6,4))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
